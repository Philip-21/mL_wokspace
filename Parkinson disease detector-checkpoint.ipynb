# Preparing the DataSet

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
lc = LabelEncoder()
df= df.apply(lc.fit_transform)
enc = OneHotEncoder()
enc.fit(df)
onehotlabel = enc.transform(df).toarray
print(df.info())
```

# creating inputs and output sets

```python

X = df.drop(['status'], axis=1) #input data
Y = df['status'] #output data
```

```python
#initialize a MinMaxScaler which scales all the data features in the range [0, 1] or else in the range [-1, 1] if there are negative values in the dataset.
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X= scaler.fit_transform(X)
Y= np.array(Y)
```

# split into train and test sets

```python
#split into train and test sets 
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(X,Y, test_size=0.2 )
```

# choose a model

```python

import xgboost as xgb
from xgboost import XGBClassifier
model = XGBClassifier()
model.fit(x_train,y_train)
```

# Finetune the model

```python
#fine tune the model using k fold crossvalidation 
from sklearn.model_selection import KFold 
newfold =KFold(12,True,model,X,Y,1)
```

# make predictions

```python
predict = model.predict(x_test)
predict
```

# calculate accuracy

```python

from sklearn.metrics import accuracy_score
accuracy_score(y_test,predict)*100
```
