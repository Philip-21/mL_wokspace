```python
import numpy as np
import pandas as pd
import pandas as pd
from pandas import DataFrame
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df = pd.read_csv('train.csv')
df
```

# adjusting trainsets

```python
df.drop(['Cabin','Embarked', 'Ticket', 'Parch'],1,inplace=True)
df
```

```python
#transforming categorical data into numerical data 
lc = LabelEncoder()
df= df.apply(lc.fit_transform)
enc = OneHotEncoder()
enc.fit(df)
onehotlabel = enc.transform(df).toarray()
print(df.info())
#creating input and outputs 
X = df.drop(['Survived'], axis = 1) #input
Y = df['Survived'] #output
print(type(X))
```

# adjsuting test sets

```python
df1 = pd.read_csv('test.csv') 
df1
```

```python
df1.drop(['Cabin','Embarked', 'Ticket', 'Parch'],1,inplace=True)
df1
```

```python
#transforming categorical data into numerical data
lc = LabelEncoder()
df1= df1.apply(lc.fit_transform)
lc2 = OneHotEncoder()
lc2.fit(df1)
onehotlabel = lc2.transform(df1).toarray()
print(df1.info())
```

```python
#CREATE A MODEL
from sklearn import svm
model = svm.SVC()
model.fit(X,Y) #TRAIN THE MODEL

predictions = model.predict(df1) #PREDICT
print(predictions)
```

```python
#CALCULATE ITS ACCURACY
#calculating accuracy using k-fold cross validation
from sklearn.model_selection import cross_val_score
train_score = cross_val_score(model, X,Y, scoring='accuracy', cv=5)
print(f'Accuracy: {(train_score*100)}%')
```
