```python
df.drop(['stab'],axis=1,inplace=True)
df
```

```python
X = df.drop(['stabf'],axis=1)
Y = df['stabf']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y, train_size=0.8, test_size=0.2, random_state=1)
from sklearn.preprocessing import MinMaxScaler
MinMaxScaler(x_train,y_train,x_test)
```

```python
#using esemble methods and multiclassifications on the dataset 
from sklearn.tree import  DecisionTreeClassifier
from xgboost import XGBClassifier 
from lightgbm import LGBMClassifier
model1 = DecisionTreeClassifier().fit(X,Y)
model2 = XGBClassifier().fit(X,Y)
model3 =  LGBMClassifier().fit(X,Y)
```

```python
model1.predict(x_test) #making predictions with the decision tree
```

```python
#accuracy of the deciscion tree model
from sklearn.model_selection import KFold 
kf =KFold(n_splits= 5)
kf.split(x_train)
from sklearn.model_selection import cross_val_score
train_score = cross_val_score(model1, X,Y, scoring='accuracy', cv=5)
print(f'Accuracy: {(train_score*100)}%')
```

```python
#accuracy of LGBM model
model2.predict(x_test)
train_score = cross_val_score(model3, X,Y, scoring='accuracy', cv=5)
print(f'Accuracy: {(train_score*100)}%')
```
